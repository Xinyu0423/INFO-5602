```python
import altair as alt
import pandas as pd 
from collections import Counter
```

```python
df2=pd.read_csv("heart_2020_cleaned.csv.csv")

df2=df2.head(500)
df2.head(10)
```

```python
male=0
female=0
male_has_disease=0
male_non_disease=0
for i in df2["Sex"]:
    if i=="Male":
        male+=1
    if i=="Female":
        female+=1

print("male:",male)
print("female:",female)


print(Counter(df2["Sex"]))


print("HeartDisease",Counter(df2["HeartDisease"]))

filterDF_Male= df2[(df2['Sex'] =="Male") & (df2['HeartDisease'] =="Yes") ]
print("Male has HeartDisease",len(filterDF_Male))

filterDF_Male= df2[(df2['Sex'] =="Female") & (df2['HeartDisease'] =="Yes") ]
print("Female has HeartDisease",len(filterDF_Male))

```

```python
data1= pd.DataFrame.from_records([
    {"a": "Male_Total","b": 159}, {"a": "Female_Total","b": 341},
    {"a": "Male_Has_Heart_disease","b": 29}, {"a": "Female_Has_Heart_disease","b": 44}
    ])

data1_Male=pd.DataFrame.from_records([
    {"a": "Male_Has_Heart_disease","b": 29},{"a": "Male_Total","b": 159}
    ])

data1_Female=pd.DataFrame.from_records([
    {"a": "Female_Has_Heart_disease","b": 44},{"a": "Female_Total","b": 341}
    ])
    
data1_Male_Pie = pd.DataFrame({"category": ["Male_Has_Heart_disease","Male_Does_Not_Has_Heart_Diesase"], "value": [29/159,1-29/159]})

data1_Female_Pie = pd.DataFrame({"category": ["Female_Has_Heart_disease","Female_Does_Not_Has_Heart_Diesase"], "value": [44/341,1-44/341]})

```

```python
print(data1_Male)
print(data1_Female)
alt.Chart(data1_Male_Pie).mark_arc().encode(
    theta=alt.Theta(field="value", type="quantitative"),
    color=alt.Color(field="category", type="nominal"),
).properties(
    title='Risk of Heart Disease in Male'
)


```

```python
alt.Chart(data1_Female_Pie).mark_arc().encode(
    theta=alt.Theta(field="value", type="quantitative"),
    color=alt.Color(field="category", type="nominal"),
).properties(
    title='Risk of Heart Disease in Female'
)
```

```python
filterDF_BMI= df2[(df2['BMI'] >30) & (df2['HeartDisease'] =="Yes") ]
print("BMI",len(filterDF_BMI))

filterDF_Smoking= df2[(df2['Smoking'] =="Yes") & (df2['HeartDisease'] =="Yes") ]
print("Smokeing",len(filterDF_Smoking))

filterDF_AlcoholDrinking= df2[(df2['AlcoholDrinking'] =="Yes") & (df2['HeartDisease'] =="Yes") ]
print("AlcoholDrinking",len(filterDF_AlcoholDrinking))

filterDF_SleepTime= df2[(df2['SleepTime']<6) & (df2['HeartDisease'] =="Yes") ]
print("SleepTime",len(filterDF_SleepTime))

filterDF_Diabetic= df2[(df2['Diabetic']=="Yes") & (df2['HeartDisease'] =="Yes") ]
print("Diabetic",len(filterDF_Diabetic))

filterDF_Asthma= df2[(df2['Asthma']=="Yes") & (df2['HeartDisease'] =="Yes") ]
print("Asthma",len(filterDF_Asthma))

filterDF_Stroke= df2[(df2['Stroke']=="Yes") & (df2['HeartDisease'] =="Yes") ]
print("Stroke",len(filterDF_Stroke))

```
